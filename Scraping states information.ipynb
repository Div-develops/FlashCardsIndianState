```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

### Request to receive state's information

```python
webpage = requests.get('https://pwonlyias.com/states-and-capitals-of-india/').text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h2'):
  print(i.text.strip())
```

```python
table_state = soup.find_all('td', attrs={'style': 'width: 26.6064%; border-style: solid; border-color: #000000;'})
new_table_state = table_state[1:]
states=[]
capital=[]
established_date=[]
official_language=[]
words=[]
for i in new_table_state:
  states.append(i.get_text())

print(states)

table_capital = soup.find_all('td',attrs={'style':'width: 32.1285%; border-style: solid; border-color: #000000;'})
new_table_capital = table_capital[1:]

for i in new_table_capital:
  capital.append(i.get_text())

print(capital)

table_established_date = soup.find_all('td',attrs={'style':'width: 12.9518%; border-style: solid; border-color: #000000;'})
new_table_established_date = table_established_date[2:]

for i in new_table_established_date:
  established_date.append(i.get_text())

print(established_date)

table_languages_spoken = soup.find_all('td',attrs={'style':'width: 16.7496%; border-style: solid; border-color: #000000;'})
new_table_languages_spoken = table_languages_spoken[2:]

for i in new_table_languages_spoken:
  words.append(i.get_text())

official_language = list(map(lambda x: x.replace('\xa0', ''), words))
print(official_language)

states_dict={'state':states,'capital':capital,'date_of_formation':established_date,'official_language':official_language}
states_df = pd.DataFrame(states_dict)
states_json = states_df.to_json(orient='records')
print('\n')
print('\n')

states_json
```

### Requests to receive Costume Images

```python
img_webpage = requests.get('https://www.myhubble.money/blog/a-visual-guide-to-the-traditional-costumes-of-all-indian-states').text
```

```python
img_soup = BeautifulSoup(img_webpage,'lxml')
```

```python
print(img_soup.prettify())
```

```python
image = img_soup.find_all('div',class_=['content-block-1', 'w-richtext'])
img_list=[]
image = image[:-2]
for i in image:
    # Find all img tags within the current element
    img_tags = i.find_all('img')

    # Iterate over each img tag
    for img_tag in img_tags:
        # Get the src attribute value (URL of the image) and append to img_list
        img_url = img_tag.get('src')
        img_list.append(img_url)
states_dict['costume_img'] = img_list
states_dict
states_df = pd.DataFrame(states_dict)
states_df
```

Converting pandas Dataframe that contains state information to JSON

```python
states_json = states_df.to_json(orient='records')

with open("states.json", "w") as f:
    f.write(states_json)
```
